```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from load_datasets import load_mnist_imgs_and_labels, load_cifar10_imgs_and_labels, load_celeba_images
```

```python
imgs, labels = load_mnist_imgs_and_labels(
  '../datasets/mnist-digits/train-images-idx3-ubyte',
  '../datasets/mnist-digits/train-labels-idx1-ubyte'
)
print(imgs.shape, labels.shape)
samples = imgs[0:16, :]
samples = samples.reshape(16, 28, 28)

fig, axes = plt.subplots(4, 4)
fig.tight_layout()

for i, ax in enumerate(axes.flatten()):
  ax.imshow(samples[i, :, :], cmap='gray')
  ax.set_title(labels[i])
  ax.set_axis_off()
```

```python
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
imgs, labels = load_cifar10_imgs_and_labels('../datasets/cifar-10-py/cifar-10-batches-py/data_batch_1')
imgs = (imgs * 255).astype(np.uint8)
print(imgs.shape, labels.shape)

for i in range(10):
	print(classes[labels[i]])
	display(Image.fromarray(imgs[i]))
```
